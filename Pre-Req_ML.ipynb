*Question 2:*

```python
def matMul(A, B):
    resultMat = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]
            
            
    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                resultMat[i][j] += A[i][k]*B[k][j]
                
    return resultMat

n, k, m = 20, 30, 20
A = [[random.random() for _ in range(k)] for _ in range(n)]
B = [[random.random() for _ in range(m)] for _ in range(k)]

start = time.time()

ans = matMul(A, B)

end = time.time()

time_without = end - start
print(f"Time taken for the multiplication of two Matrices A and B without using numpy is: {time_without} s")

print()


for row in ans:
    print(row)
```

```python
A1 = np.array(A)
B1 = np.array(B)

start = time.time()
ans_np = np.dot(A1, B1)
end = time.time()

print(f"Time taken for the multiplication of two Matrices A and B with using numpy is: {end - start} s")

print()


for row in ans_np:
    print(row)

```

Using Numpy is faster than multiplying matrices using three for loops because NumPy operations are vectorized means that they operate on entire arrays at once using optimized libraries.

```python
def find_median_1(arr):
    n = len(arr)
    
    for i in range(n // 2 - 1):
        max_elem = max(arr)
        arr.remove(max_elem)
    
    max1 = max(arr)
    arr.remove(max1)
    max2 = max(arr)
    
    if n % 2 == 0:
        median = (max1 + max2) / 2
    else:
        median = max2
    
    return median

np_arr = np.random.randint(0, 100, size=1000)
arr = list(np_arr)

start_time = time.time()

median = find_median_1(arr)

end_time = time.time() 

print(f"The Median is {median} by one pass method.")    
print(f"Time taken by the above method is {end_time - start_time}")
```

Time Complexity of above method is O(n^2)

```python
def find_median_2(arr):
    arr.sort()
    n = len(arr)
    
    if n % 2 == 0:
        median = (arr[n//2 - 1] + arr[n//2]) / 2
    else:
        median = arr[n//2]
        
    return median

np_arr = np.random.randint(0, 100, size=1000)
arr = list(np_arr)


start_time = time.time()

median = find_median_2(arr)

end_time = time.time()

print(f"The Median is {median} by sorting method.")  
print(f"Time taken by the above method is {end_time - start_time}")
```

Time Complexity of above method is O(nlog(n))

*Question 8:*

```python
class Student:
    def __init__(self, name, roll_number, marks):
        self.name = name
        self.roll_number = roll_number
        self.marks = marks

class Branch:
    def __init__(self, name):
        self.name = name
        self.students = []

    def add_student(self, name, roll_number, marks):
        self.students.append(Student(name, roll_number, marks))

class Year:
    def __init__(self, year):
        self.year = year
        self.branches = []

    def add_branch(self, name):
        self.branches.append(Branch(name))

class Database:
    def __init__(self):
        self.years = []

    def add_year(self, year):
        self.years.append(Year(year))
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.5, 100.5, 0.5)

plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.plot(x, x, label='y=x', color='blue')
plt.title("1. y = x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

plt.subplot(2, 3, 2)
plt.plot(x, x**2, label='y=x^2', color='green')
plt.title("2. y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

plt.subplot(2, 3, 3)
plt.plot(x, (x**3)/100, label='y=(x^3)/100', color='red')
plt.title("3. y = (x^3)/100")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

plt.subplot(2, 3, 4)
plt.plot(x, np.sin(x), label='y=sin(x)', color='purple')
plt.title("4. y = sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

plt.subplot(2, 3, 5)
plt.plot(x, np.sin(x)/x, label='y=sin(x)/x', color='orange')
plt.title("5. y = sin(x)/x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

plt.subplot(2, 3, 6)
plt.plot(x, np.exp(x), label='y=exp(x)', color='cyan')
plt.title("6. y = e^x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

plt.tight_layout()

plt.show()
```
